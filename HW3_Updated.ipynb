```python
%matplotlib notebook

import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Surface
Y = np.arange(0, 100, 0.1)
X = np.arange(0, 100, 0.1)

#Constraint
Y_bar = np.arange(40, 100, 0.1)
X_bar = 200 - 2*Y_bar

X, Y = np.meshgrid(X, Y)
Z = 50*np.multiply(np.power(X, 2),np.power(Y,2))

Z_bar = 50*np.multiply(np.power(X_bar, 2),np.power(Y_bar,2))
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('P(X,Y)')

# Plot surface
surf = ax.plot_surface(X, Y, Z, linewidth=0, label = 'Function P(X,Y)', zorder=0)

#PLot constraint and optimal soln
ax.plot(X_bar, Y_bar, Z_bar, label = 'Constriant', linewidth = 4, zorder=1)
ax.scatter3D(80, 60, 50*np.multiply(np.power(80, 2),np.power(60,2)),c = 'r', label = 'Point (80, 60)', s = 60, zorder=2)

surf._facecolors2d=surf._facecolors3d
surf._edgecolors2d=surf._edgecolors3d
ax.legend()
#ax.show()
```

```python
fig1 = plt.figure()
ax1 = Axes3D(fig1)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

#Q2
X = np.arange(5,15,0.5)
Y = np.arange(0,10,0.5)
Z = np.arange(0,5,0.5)
XX, YY, ZZ = np.meshgrid(X, Y, Z)
M = np.multiply(XX, YY)
MM = np.multiply(M, ZZ)

img1 = ax1.scatter(XX, YY, ZZ, c=MM.flatten(), cmap="RdPu", alpha=0.5, label = 'Objective Function (Color)')
fig1.colorbar(img1)
#plt.show()

#Constraint plane(surface)
XA = np.arange(8,15,0.5)
YA = np.arange(2,10,0.5)
XC, YC = np.meshgrid(XA, YA)
ZC = (96 - 3*XC - 4*YC)/12
surf = ax1.plot_surface(XC, YC, ZC, label = 'Constraint Surface')
surf._facecolors2d=surf._facecolors3d
surf._edgecolors2d=surf._edgecolors3d

# Optimal Soln
ax1.scatter3D(10.6667, 8, (96 - 3*10.6667 - 4*8)/12,c = 'r', label = 'Optimal Soln (10.6667,8,2,6667)', s = 60)
ax1.legend()
```
